```python
def IsInPolygon(data_infrastructure_inp, polygons_inp):#функция для проверки вхождения в область
  result = []
  for data in data_infrastructure_inp:
    result.append(0)
    stop = False
    for element in data['features']:
      element = element['geometry']['coordinates']
      try:
        point = Point(element[1],element[0])
      except:
        point = Point(element[0][1],element[0][0])
      for i in range(3):
        if (polygons_inp[i].contains(point)):
          result[-1] = i + 1
          if (i == 0):
            stop = True
          break
      if (stop):
        break
  return result
```

```python
def changeData(data_inp):#преобразуем коэффициент в время
  result = []
  for d in data_inp:
    if (d == 1):
      result.append(5)
    elif (d == 2):
      result.append(15)
    elif (d == 3):
      result.append(30)
    else:
      result.append(0)
  return result
```

```python
def changeData3(data_inp): # оформляем время в текстовом виде
  result = []
  for d in data_inp:
    if (d == 0):
      result.append(45)
    else:
      result.append(d)
  return result
```

```python
def changeData2(data_inp): # оформляем время в текстовом виде
  result = []
  for d in data_inp:
    if (d == 0):
      result.append("Рядом нет")
    else:
      result.append("До "+ str(d) + " минут пешком")
  return result
```

```python
def eval_years(x):#оценка целевого возраста
  a1 = 30 # средний возраст людей с детьми (посетители детских площ.)
  a2 = 27 # средний возр. людей, посещающих тренажёрные залы
  a3 = 60 # средний возраст пациентов больниц для взрослых
  a4 = 39 # средний возраст родителей с детьми до 18 лет (детские больницы)
  a5 = 30
  a6 = 44
  S = int ((a1*x[0] + a2*x[1] + a3*x[3] + a4*x[4] + a5*x[6] + a6*x[7])/(sum(x) - x[2] - x[5] + 0.00001))
  return S
```

```python
def calculateColor(data_inp):#окрашиваем объект на основе возраста
  if (data_inp < 35):
    return "blue"
  elif (data_inp < 38):
    return "purple"
  elif (data_inp < 40):
    return "yellow"
  else:
    return "red"
```

```python
def eval_rating(x):
  S = sum(x)
  if (S == 0):
    return 0
  else:
    return (5.5 - S/80)
```
